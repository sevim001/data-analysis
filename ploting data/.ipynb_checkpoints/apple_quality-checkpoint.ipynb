```python
import math
import collections

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
pd.options.display.max_rows = 16
```

```python
apple = pd.read_csv('apple_quality.csv')
```

```python
apple.head()
```

```python
apple.describe()
```

```python
apple_by_quality = apple.set_index("Quality")
```

```python
apple_by_quality.loc["good"].plot.scatter("Weight","Sweetness")
```

```python
apple_by_quality.loc["good"].sort_values("Weight").plot("Weight","Sweetness")
```

```python
axes = apple_by_quality.loc["good"].sort_values("Weight").plot("Weight","Sweetness")
axes = apple_by_quality.loc["bad"].sort_values("Weight").plot("Weight","Sweetness", ax=axes)
```

```python
axes = apple_by_quality.loc["good"].sort_values("Size").plot("Size","Sweetness")
axes = apple_by_quality.loc["bad"].sort_values("Size").plot("Size","Sweetness", ax=axes)
```

```python
axes = apple_by_quality.loc["good"].sort_values("Ripeness").plot("Ripeness","Sweetness")
axes = apple_by_quality.loc["bad"].sort_values("Ripeness").plot("Ripeness","Sweetness", ax=axes)
```
