```python
import math
import collections

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
car = pd.read_csv("file.csv")
```

```python
car.head()
```

```python
#data cleaning
car.isnull().sum()
```

```python
car["Cylinders"].fillna(car["Cylinders"].mean , inplace = True) 
```

```python
#show all the records when the region is asia or europe
car[car["Origin"].isin(["Asia" , "Europe"])]
```

```python
#remove all the records where weight is above 400
car[~(car["Weight"] > 4000)]
```

```python
#increase all the values of MPG_City by 3
car["MPG_City"] = car["MPG_City"].apply(lambda x:x+3)
```

```python
car
```

```python
car_origin = car.set_index("Origin").sort_index()
```

```python
car_origin.loc["USA"].sort_values("MSRP").plot("MSRP","Horsepower")
#the relation between the car price and horsepower in usa
```

```python
axes = car_origin.loc["USA"].sort_values("MSRP").plot("MSRP","Horsepower" ,label="USA")
car_origin.loc["Asia"].sort_values("MSRP").plot("MSRP","Horsepower" ,ax=axes ,label="Asia")
car_origin.loc["Europe"].sort_values("MSRP").plot("MSRP","Horsepower",ax=axes ,label="Europe")
pp.ylabel('Horsepower')
```
