```python
data['Salary Estimate'] = data['Salary Estimate'].str.replace('K','')
```

```python
data['Salary Estimate'] = data['Salary Estimate'].str.replace('(Glassdoor est.)','')
```

```python
data.head(5)
```

```python
data[["max_salary" , "min_salary"]] = data["Salary Estimate"].str.split('-' ,expand = True )
```

```python
data.head(5)
```

```python
data['Job Description'].unique
```

```python
data['max_salary'] = data['max_salary'].str.replace('$','')
data['max_salary'] = data['max_salary'].astype(int)
data.max_salary.max()
```

```python
#finding the highest paid job describtion
data[data['max_salary'] == 212]['Job Description']
```

```python
data['avg_salery']= (data['max_salary']+data['min_salary'])/2
```

```python
data['min_salary'] = data['min_salary'].str.replace('$','')
data['min_salary'] = data['min_salary'].str.replace('(Employer est.)','')
data['min_salary'] = data['min_salary'].astype(int)

data.min_salary.min()
```
